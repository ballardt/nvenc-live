```python
# Just hardcoded before presentation
import pandas as pd
import seaborn as sb

data = {'eval': ['time', 'time', 'time', 'time', 'time', 'size', 'size', 'size', 'size', 'size', 'qualSize', 'qualSize'],
        'config': ['1x3', '3x1', '3x3', '5x5', '9x9', '1x2', '2x1', '2x2', '4x4', '8x8', 'high', 'low'],
        'time_s': [20.902, 20.786, 22.910, 22.635, 23.773, 22.860, 22.861, 22.486, 22.750, 23.571, 26.880, 19.007],
        'size_kb': [5517, 5517, 6464, 6335, 6741, 6478, 6478, 6155, 6422, 6741, 8518, 4370]}
df = pd.DataFrame(data)
df['size_mb'] = df['size_kb'] / 1000
```

```python
# Time vs tiles
bplt = sb.barplot(x='config', y='time_s', data=df[df['eval']=='time'])
bplt.set_ylim((0, 32))
bplt.axhline(y=30, linestyle='--', color='red')
bplt.set_title("Number of tiles vs. encoding time")
bplt.set_ylabel("time [s]")
bplt.set_xlabel("tile configuration")
```

```python
# Filesize vs tiles
bplt = sb.barplot(x='config', y='size_mb', data=df[df['eval']=='size'])
bplt.set_title("Number of tiles vs. output size")
bplt.set_ylabel("output file size [MB]")
bplt.set_xlabel("tile configuration")
```

```python
# Quality vs time
bplt = sb.barplot(x='config', y='time_s', data=df[df['eval']=='qualSize'])
bplt.set_title("4x4, all high or low vs. encoding time")
bplt.set_ylabel("time [s]")
bplt.set_xlabel("tile quality")
bplt.set_ylim((0, 32))
bplt.axhline(y=30, linestyle='--', color='red')
```
