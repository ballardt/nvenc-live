```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline

# Load the CSV data
df = pd.read_csv('ssim_sample.csv', header=None)
df.columns = ['qual', 'tiles', 'ssim']

# Configure the data
TILES = ['1x1', '1x2', '2x2']
df = df[df['tiles'].isin(TILES)]
high_df = df[df['qual']=='high']
mixed_df = df[df['qual']=='mixed']
low_df = df[df['qual']=='low']
```

```python
# Matplotlib params
plt.rcParams['font.family'] = 'serif'
```

```python
# Plot SSIM
ax = plt.gca()
high_df.plot(x='tiles', y='ssim', label='all high', linewidth=2, 
             linestyle='--', ax=ax)
mixed_df.plot(x='tiles', y='ssim', label='mixed', linewidth=2, 
              linestyle='-', ax=ax)
low_df.plot(x='tiles', y='ssim', label='all low', linewidth=2, 
            linestyle='-.', ax=ax)

# Style
plt.xticks(np.arange(len(df['tiles'].unique())), df['tiles'].unique())
plt.grid(True, linestyle=':', linewidth=2)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlim(0, len(TILES)-1)
ax.set_ylim(0.5, 1)

plt.show()
```
